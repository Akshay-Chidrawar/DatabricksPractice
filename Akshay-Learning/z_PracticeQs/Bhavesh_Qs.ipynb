- **show only those IDs** for which the Sales is strictly increasing **for every YOY**.
- **show all those IDs** for which the Sales has increased **for at least one YOY**.

```sql
use catalog workspace;
create database if not exists bhavesh_qs;
use workspace.bhavesh_qs;
```

```sql
CREATE TABLE if NOT EXISTS sales (
  Year INT,
  ID INT,
  Sales INT
);

INSERT INTO sales VALUES
(2014, 1, 10),
(2015, 1, 5),
(2016, 1, 15),
(2014, 2, 20),
(2015, 2, 25),
(2016, 2, 10),
(2014, 3, 10),
(2015, 3, 20),
(2016, 3, 30);
```

```sql
select * from workspace.bhavesh_qs.sales;
```

```sql
create or replace temporary view vw_sales as
select *
    ,lag(Sales) over (partition by ID order by Year) as prev_sales
    ,Sales-prev_sales as diff
from  workspace.bhavesh_qs.sales;

select * from vw_sales;
```

```sql
select distinct id from vw_sales
where ID not in (select ID from vw_sales where diff<0)
```

```sql
select distinct a.id from vw_sales a
where a.diff >0
```

### Where Exists
- Enlist all suppliers who have at least one product cheaper than 20.

```sql
CREATE OR REPLACE TABLE products (
    ProductID INT,
    ProductName STRING,
    SupplierID INT,
    Price INT
);

INSERT INTO products VALUES
    (1, 'Chais', 1, 18),
    (2, 'Chang', 1, 19),
    (3, 'Aniseed Syrup', 1, 10),
    (4, 'Chef Anton''s Cajun Seasoning', 2, 22),
    (5, 'Chef Anton''s Gumbo Mix', 2, 30);

CREATE OR REPLACE TABLE suppliers (
    SupplierID INT,
    SupplierName STRING
);

INSERT INTO suppliers VALUES
    (1, 'Exotic Liquid'),
    (2, 'New Orleans Cajun Delights'),
    (3, 'Grandma Kelly''s Homestead'),
    (4, 'Tokyo Traders');
```

```sql
select * from products;
```

```sql
select s.* from workspace.bhavesh_qs.suppliers s
where exists (
  select p.ProductID
  from workspace.bhavesh_qs.products p
  where p.Price<20 and s.SupplierID = p.SupplierID
  )
```

```sql
select distinct s.SupplierID,s.SupplierName 
from workspace.bhavesh_qs.suppliers s
inner join workspace.bhavesh_qs.products p on s.SupplierID = p.SupplierID and p.Price<20
```
