> data = json.loads(response.text)

- **accepts JSON file object (f) as input**
1. **load() --> reads data from "f" into a new dict/ list**
2. **dump() --> writes data from existing dict/ list of dict into "f"**
- **accepts string (s) as input**
1. **loads() --> reads data from "s" into a new dict/ list**
2. **dumps() --> writes data from existing dict/ list of dict into "s"**

```python
directory = home_dir+'coffee_sales/'
output_file = directory+'output.json'
combined_data = []

for filename in os.listdir(directory):
    if filename.endswith('.json') and filename != 'coffee_sales.json' and filename != 'output.json':
        with open(os.path.join(directory, filename), 'r') as f:
            data = json.load(f)
            combined_data.extend(data)

with open(output_file, 'w') as f:
    json.dump(combined_data, f)

#list of dictionaries
print(combined_data,"\n")

#individual dictionaries
for item in combined_data:
    print(item)

with open(output_file, 'r') as f:
    data = json.load(f)
    for item in data:
       print(item)
```
