```python
filename='/kaggle/working/writefile.txt'
try: 
    #os.path.exists('/kaggle/working/writefile.txt')
    os.remove(filename)
    print('existing file removed')
except:
    pass

f = open(filename,'x')  #creates new file, only if not already exists. else, it gives error.
f.write('create file if not exists')
f.close()

f = open(filename)
print(f.read())
f.close()

f = open(filename,'w') #creates new file, irresepctive of weather it already exists (overwrite) or not.
f.write('overwrite previous file')
f.close()

f = open(filename)
print(f.read())
f.close()
```

```python
import csv
filename = '/kaggle/input/samplecsv/sample.csv'

with open(filename) as f:
    
    csv_reader = csv.reader(f)
    
    for i,row in enumerate (csv_reader):
        if i==0:
            print(f"{'|'.join(row)}")
        else:
            print(row)

    print('\n')
   
```

```python
filename = '/kaggle/input/samplecsv/sample.csv'

with open(filename) as f:
    
    dict_reader = csv.DictReader(f)
    
    for i,row in enumerate (dict_reader):
        if i==0:
            print(f"{' '.join(row)}")
        else:
            print(row)
```

```python
filename = '/kaggle/working/writecsv.csv'

with open(filename,'w') as f:    
    csv_writer = csv.writer(f)
    csv_writer.writerow(["4","d"])
    csv_writer.writerow(["5","e"])

f = open(filename)
print(f.read())
```

```python
filename = '/kaggle/working/writecsv.csv'

with open(filename,'w') as f:    
    dict_writer = csv.DictWriter(f,fieldnames=["eid","ename"])

    dict_writer.writeheader()
    dict_writer.writerow({"eid":"4","ename":"d"})
    dict_writer.writerow({"ename":"e","eid":"5"})

f = open(filename)
print(f.read())
```

```python

values=[{"eid":"4","ename":"d"},{"ename":"e","eid":"5"}]

filename = '/kaggle/working/writecsv.csv'

with open(filename,'w') as f:    
    dict_writer = csv.DictWriter(f,fieldnames=values[0].keys(),dialect="excel")

    dict_writer.writeheader()
    dict_writer.writerow({"eid":"6","ename":"f"})
    dict_writer.writerow({"ename":"e","eid":"5"})

f = open(filename)
print(f.read())
```
